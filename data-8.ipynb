```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('D:/GC/age.csv')
data6= data.loc[:,['Diameter','Height','Whole_Weight','Viscra_Weight','Shell_Weight','Group']]
data6.loc[data['Group']==0,['Group']]='Year0'
data6.loc[data['Group']==1,['Group']]='Year1'
data6.loc[data['Group']==2,['Group']]='Year2'
data6
```

```python
data6.columns
```

```python
sns.countplot(data =data6, x="Group")
plt.xlabel("Cancer or Not")
plt.ylabel("Count")
plt.title("Label")
```

```python
Year0=data6.loc[data6.Group=='Year0',['Diameter', 'Height', 'Whole_Weight', 'Viscra_Weight', 'Shell_Weight']]
Year1=data6.loc[data6.Group=='Year1',['Diameter', 'Height', 'Whole_Weight', 'Viscra_Weight', 'Shell_Weight']]
Year2=data6.loc[data6.Group=='Year2',['Diameter', 'Height', 'Whole_Weight', 'Viscra_Weight', 'Shell_Weight']]

Year0
```

```python
plt.figure(figsize=(10,6))

sns.histplot(Year0['Diameter'],bins=7,alpha=0.5,label='Year0',kde=True)
sns.histplot(Year1['Diameter'],bins=7,alpha=0.5,label='Year1',kde=True)
sns.histplot(Year2['Diameter'],bins=7,alpha=0.5,label='Year2',kde=True)

plt.xlim(0,1)
plt.xlabel('DIameter')
plt.ylabel('Count')
plt.title('Waist Seaborn Histogram - KDE(Kernel Density Estimator)')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(Year0['Height'],bins=7,alpha=0.5,label='Year0',kde=True)
sns.histplot(Year1['Height'],bins=7,alpha=0.5,label='Year1',kde=True)
sns.histplot(Year2['Height'],bins=7,alpha=0.5,label='Year2',kde=True)

plt.xlim(0,0.5)
plt.xlabel('Height')
plt.ylabel('Count')
plt.title('Waist Seaborn Histogram - KDE(Kernel Density Estimator)')
plt.legend()
plt.grid()
plt.show()
```

```python
color = ['#00994C', '#FF007F','b']
pp = sns.PairGrid(data6,hue='Group',palette=color,height=3.3,aspect=1.3)
pp.map_diag(sns.histplot,bins=10)
pp.map_offdiag(sns.kdeplot)
pp.add_legend()
```

```python
new_df=data6.loc[:,['Diameter','Height','Group']]
new_df.loc[data['Diameter']>=0.4,['Diameter']]='BIG'
new_df.loc[data['Diameter']<0.4,['Diameter']]='SMALL'
new_df
```

```python
Diameter = new_df.groupby(['Group','Diameter'])['Diameter'].count()

Diameter = Diameter.to_frame(name='count')

Diameter = Diameter.reset_index()
```

```python
plt.figure(figsize=(10, 6))

# 막대 그래프 생성
sns.barplot(x='Group', y='count', hue='Diameter', data=Diameter)

# X, Y 축 레이블 설정
plt.xlabel('Group')
plt.ylabel('Count')

# 그래프 제목 설정
plt.title('Count of BIG and SMALL by Group')

# 범례 추가
plt.legend(title='Diameter')

# 그리드 추가
plt.grid()

# 그래프 표시
plt.show()
```
